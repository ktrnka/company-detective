```python
import re
from typing import Optional

def extract_apple_app_store_id(url: str) -> Optional[int]:
    """Get the Apple App Store ID from an app store URL, or return None if not found."""
    pattern = r'https://apps.apple.com/us/app/[^/]+/id(\d+)'
    match = re.search(pattern, url)
    if match:
        return int(match.group(1))
    else:
        return None
    
extract_apple_app_store_id("https://apps.apple.com/us/app/98point6/id1157653928"), extract_apple_app_store_id("https://www.98point6.com/press_release/98point6-now-available-nationwide/")
```

```python
from urllib.parse import urlparse, parse_qs

def extract_google_play_app_id(url: str) -> Optional[str]:
    """Get the Google Play Store app ID from an app store URL, or return None if not found."""
    parsed_url = urlparse(url)
    if parsed_url.netloc == 'play.google.com' and parsed_url.path == '/store/apps/details':
        query_params = parse_qs(parsed_url.query)
        app_id = query_params.get('id')
        if app_id:
            return app_id[0]
    return None

extract_google_play_app_id("https://play.google.com/store/apps/details?id=com.ninety8point6.patientapp&hl=en_US"), extract_google_play_app_id("https://www.98point6.com/press_release/98point6-now-available-nationwide/")
```

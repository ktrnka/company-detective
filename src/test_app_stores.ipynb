```python
from core import CompanyProduct, init

init()
```

```python
target = CompanyProduct("Singularity 6", "Palia")
```

```python
import app_stores.steam as steam

url = steam.find_steam_page(target)
url
```

```python
steam_id = steam.extract_steam_id(url)
steam_id
```

```python
# This will be near-instant if cached! Otherwise about 5 sec
reviews = steam.get_reviews(steam_id)
reviews
```

```python
for review in reviews:
    print(steam.review_to_markdown(review), "\n\n")
```

```python
review_summary_stats = steam.get_review_summary_stats(steam_id)
review_summary_stats
```

```python
print(steam.summarize_sampling(reviews, review_summary_stats))
```

# Google Play

```python
import app_stores.google_play as google_play

target = CompanyProduct.same("98point6")

url = google_play.find_google_play_page(target)
google_play_id = google_play.extract_google_play_app_id(url)
google_play_id
```

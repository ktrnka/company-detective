# Apple app store

```python
import app_stores.apple as apple_app_store
from core import Seed

target = Seed.init("98point6")

url = apple_app_store.find_app_store_page(target)
apple_app_store_id = apple_app_store.extract_apple_app_store_id(url)
apple_app_store_id
```

```python
apple_reviews = apple_app_store.scrape(apple_app_store_id)
apple_reviews
```

```python
for review in apple_reviews:
    print(apple_app_store.review_to_markdown(review).strip(), "\n\n")
```

# Join and summarize!

```python
from llm_utils import pack_documents


review_markdowns = [apple_app_store.review_to_markdown(review) for review in apple_reviews] + [google_play.review_to_markdown(review) for review in google_play_reviews]

print("Total reviews:", len(review_markdowns))

packed_reviews = pack_documents(review_markdowns, max_chars=50000)

print("Packed reviews:", len(packed_reviews))
print("Lengths:", [len(doc) for doc in packed_reviews])
```
